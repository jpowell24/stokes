```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc

import math
```

```python
start_time = 0; 
total_time = 3.33; # Total time plotted in seconds
cadance = 0.3; # Frequency of stepps in hertz
period = round(1 / cadance, 2); # Period of stepping

data_points = 50; 

x, y = np.linspace(start_time, start_time + total_time, data_points, endpoint=False), np.zeros(data_points); 

u = 0.15 * ((x % period) - period / 2); 
v = 0.75 + 0.25 * np.sin(1.5 * ((2 * math.pi) / period) * x); 


rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
rc('text', usetex=True)

plt.quiver(x, y, u, v, scale = 1.35, width = 0.003); 

plt.quiver(0.3, 0, 0, 0.9, scale = 1.35, width = 0.006, color = "blue"); 
plt.quiver(0.3, 0.005, -0.15, 0, scale = 1.35, width = 0.006, color = "red"); 

plt.ylim(0,1.1)
plt.xlim(-2,5.15)

plt.tick_params(
    axis='x',          
    which='both',    
    bottom=False,     
    top=False,         
    labelbottom=False) 

plt.xlabel("Time (arb. units)")
plt.ylabel("Force (body weights)")

plt.text(0.38, 0.95, 'Vertical \n component', fontsize = 10, color = "blue")
plt.text(-1.2, 0.04, 'Horizontal \n component', fontsize = 10, color = "red")

plt.title("Ground Reaction Forces")

plt.rcParams['savefig.dpi'] = 300; 
plt.savefig(f"../comp_bio_models/temp_plot");   

plt.show() 
```
